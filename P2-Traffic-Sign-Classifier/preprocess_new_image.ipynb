# Preprocess new images 
- I took new images from videos below by taking screen shots of them.
  - https://www.youtube.com/watch?v=S_rfYOQdpwQ
  - https://www.youtube.com/watch?v=9iYligjjaa8
- In this notebook, manually preprocessing the new images so that my model can train on them.
  - trimming unnecessary parts of the new images
  - reshaping to (32, 32, 3)

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import scipy.misc
```

```python
# Difine some constants and functions
resize_to = (32, 32)

def plot_images_of(original_image, trimmed_image):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(original_image)
    ax1.set_title('Original')
    ax2.imshow(trimmed_image)
    ax2.set_title('Trimmed')
    plt.show()
```

## Prepare

```python
img_1 = mpimg.imread('./data/new_image/1.png')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img_1)
ax2.imshow(img_1[30:150, 40:160, :3])
plt.show()
```

## Trim and resize then save
- Below I trimed the original images and resize them to (32, 32, 3)

```python
n = 1
original = mpimg.imread('./data/new_image/{}.png'.format(n))
trimmed = img_1[30:150, 40:160, :3]
plot_images_of(original, trimmed)
resized = scipy.misc.imresize(trimmed, resize_to)
mpimg.imsave('./data/new_image/preprocessed_{}.png'.format(n), resized)
```
